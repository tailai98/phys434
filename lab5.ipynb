```python
import pickle
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import scipy
from scipy import stats
```

```python
infile1 = open ("higgs_100000_pt_250_500.pkl",'rb')
dictH = pickle.load(infile1)
infile2 = open ("qcd_100000_pt_250_500.pkl",'rb')
dictQ = pickle.load(infile2)
```

```python
dictH.keys()
```

The data was simulated based on real data from the ATLAS Experiment at the Large Hadron Collider at sqrt(s) = 13 TeV. The data also included probable Higgs boson data as well as the background QCD data. The data itself is measurements of the jets that are emitted during proton-proton collisions. These jets are made up of beams of particles are are a result of the high speed proton-proton collison.

```python
plt.hist(dictQ['pt'],bins=100,density=False,alpha=.5,label='QCD');
plt.hist(dictH['pt'],bins=100,density=False,alpha=.5,label='Higgs');
plt.title('Pt histograms')
plt.xlabel('pt')
plt.ylabel('Counts')
plt.legend();
```

The first variable graphed is the transverse momentum (pt) of the jets that ATLAS measures. As you can see the higgs data has a large spike in counts around the pt=475 mark, which is not seen on the background histogram. Also both of these distributions follow a rayleigh distribution, since the equation for the pt variable here is sqrt(px^2+py^2).

```python
plt.subplots(1,2)
plt.suptitle('Mass Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)
plt.subplot(1,2,1)
plt.title('QCD Mass')
plt.xlabel('Mass')
plt.ylabel('Counts')
plt.hist(dictQ['mass'],bins=800,density=False,alpha=.5);
plt.subplot(1,2,2)
plt.title('Higgs Mass')
plt.xlabel('Mass')
plt.ylabel('Counts')
plt.hist(dictH['mass'],bins=500,density=False,alpha=.5);
```

The histograms above show the mass distributions for the simulated data. The mass value was calculated using the measured energy of the jet and the momentum, which can be see as E^2 = pt^2 + M^2. The QCD histogram seems to follow a rayleigh distribution, while the Higgs histogram does not seem to follow any immediately recognizable distribution. What is notable about the Higgs plot is the massive spike in counts at approximately M=125.

```python
plt.subplots(1,2)
plt.suptitle('Eta Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)
plt.subplot(1,2,1)
plt.title('QCD Eta Value')
plt.xlabel('Eta')
plt.ylabel('Count Probability')
plt.hist(dictQ['eta'],bins=200,density=True,alpha=.5);
plt.subplot(1,2,2)
plt.title('Higgs Eta value')
plt.xlabel('Eta')
plt.ylabel('Count Probability')
plt.hist(dictH['eta'],bins=200,density=True,alpha=.5);
```

The histograms shown above describe the eta value in the experiment, which is pseudorapidity. This value represents the angle from the beam at which the jets are created and seen by the detector. A small eta value corresponds with being close to perpendicular from the proton beam and a large eta value corresponds with a smaller angle between the jet and the proton beam. Seen clearly in both histograms is a normal distribution, although the Higgs eta value plots shows a smaller standard deviation than the QCD background plot.

```python
plt.subplots(1,2)
plt.suptitle('Phi Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)
plt.subplot(1,2,1)
plt.title('QCD Phi Value')
plt.xlabel('Phi')
plt.ylabel('Count Probability')
plt.hist(dictQ['phi'],bins=200,density=True,alpha=.5);
plt.subplot(1,2,2)
plt.title('Higgs Phi value')
plt.xlabel('Phi')
plt.ylabel('Count Probability')
plt.hist(dictH['phi'],bins=200,density=True,alpha=.5);
```

These are the histograms for the phi value. The phi value corresponds to the angle between the x and y axis that the jet follows. This is different than the eta value by signifying the transverse angle, not the angle from the jet and the beam. Notice how there is not much difference between count probabilities as the phi value changes, thus meaning that this variable does not correlate with any other variable in the experiment.

```python
plt.subplots(1,3)
plt.suptitle('EE2, EE3 and d_2 Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)
plt.subplot(1,3,1)
plt.title('EE2 Values')
plt.xlabel('EE2')
plt.ylabel('Counts')
plt.hist(dictQ['ee2'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['ee2'],bins=200,density=False,alpha=.5,label='Higgs');
plt.legend()
plt.subplot(1,3,2)
plt.title('EE3 values')
plt.xlabel('EE3')
plt.ylabel('Counts')
plt.hist(dictQ['ee3'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['ee3'],bins=200,density=False,alpha=.5,label='Higgs');
plt.xlim(0,.01);
plt.legend();
plt.subplot(1,3,3)
plt.title('d_2 values')
plt.xlabel('d_2')
plt.ylabel('Counts')
plt.hist(dictQ['d2'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['d2'],bins=200,density=False,alpha=.5,label='Higgs');
plt.xlim(0,45)
plt.legend();
```

These histograms represent the 2 and 3-point ECF ratios. ECF stands for energy-correlation functions. The ratio value for the QCD background is notably closer to zero for the majority of the data, while the Higgs ratio values are much smaller in counts but further away from zero than the QCD. As expected, the d_2 histograms mirror the EE2 and EE3 values, because d_2 is a ratio of those two variables.

```python
plt.subplots(1,2)
plt.suptitle('Angularity Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)
plt.subplot(1,2,1)
plt.title('QCD Angularity Value')
plt.xlabel('Angularity')
plt.ylabel('Counts')
plt.hist(dictQ['angularity'],bins=200,density=False,alpha=.5);
plt.subplot(1,2,2)
plt.title('Higgs Angularity value')
plt.xlabel('Angularity')
plt.ylabel('Counts')
plt.hist(dictH['angularity'],bins=200,density=False,alpha=.5);
```

The histograms shown above correlate to the angularity measurements. The angularity measurement is a measurement of the radiation of the jet. The larger the angularity, the more radiation in the jet. As seen in the Higgs plot, there is a large spike in counts at approximately angularity = .05 and then an exponential decay to 0 counts at around ang=.03, whereas in the QCD plot there is a smaller spike at around the same x value but then a less steep exponential decay down to 0 counts at ang=.03.

```python
plt.subplots(1,3)
plt.suptitle('t value Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)

plt.subplot(1,3,1)
plt.title('t1 Values')
plt.xlabel('t1')
plt.ylabel('Counts')
plt.hist(dictQ['t1'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['t1'],bins=200,density=False,alpha=.5,label='Higgs');
plt.legend();
plt.subplot(1,3,2)
plt.title('t2 Values')
plt.xlabel('t2')
plt.ylabel('Counts')
plt.hist(dictQ['t2'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['t2'],bins=200,density=False,alpha=.5,label='Higgs');
plt.legend();
plt.subplot(1,3,3)
plt.title('t3 Values')
plt.xlabel('t3')
plt.ylabel('Counts')
plt.hist(dictQ['t3'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['t3'],bins=200,density=False,alpha=.5,label='Higgs');
plt.legend();
```

As seen above, the time variable histograms for the Higgs data are all similar distribution shapes, and centered relatively close to zero compared to the QCD time variables, except for t1. The distributions for the QCD times change for the increasing variables. This change is a smaller number of counts per t value, but approaching an even distribution. The Higgs distribution changes count value but not overall shape for the increasing t values.

```python
plt.subplots(1,2)
plt.suptitle('T value ratio Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)

plt.subplot(1,2,1)
plt.title('t21 Values')
plt.xlabel('t21')
plt.ylabel('Counts')
plt.hist(dictQ['t21'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['t21'],bins=200,density=False,alpha=.5,label='Higgs');
plt.legend();
plt.subplot(1,2,2)
plt.title('t32 Values')
plt.xlabel('t32')
plt.ylabel('Counts')
plt.hist(dictQ['t32'],bins=200,density=False,alpha=.5,label='QCD');
plt.hist(dictH['t32'],bins=200,density=False,alpha=.5,label='Higgs');
plt.legend();
```

These plots represent the ratios between the t2 and t1 values on the left, and the ratio between the t3 and t2 values on the right. There is a noticable difference in the t21 ratio for the higgs data, it is skewed to the left and steeper while the QCD data is smoother and skewed to the right. The t32 ratio is very similar for the two data sets, with the small difference being the number of counts for the QCD set is slightly higher than the Higgs set for ratio values greater than .7. With that difference then comes the counts for the Higgs ratios are slightly higher for values smaller than .7.

```python
plt.subplots(1,2)
plt.suptitle('KtDeltaR Histograms',x=.80,y=1.15)
plt.subplots_adjust(left=.125,right=1.5,top=1,bottom=.2,
                   wspace=.3,hspace=.01)
plt.subplot(1,2,1)
plt.title('QCD KtDeltaR')
plt.xlabel('KtDeltaR')
plt.ylabel('Counts')
plt.hist(dictQ['KtDeltaR'],bins=500,density=True,alpha=.5);
plt.subplot(1,2,2)
plt.title('Higgs KtDeltaR')
plt.xlabel('KtDeltaR')
plt.ylabel('Counts')
plt.hist(dictH['KtDeltaR'],bins=500,density=True,alpha=.5);
```

The KtDeltaR variable corresponds with the distance between jets when they are picked up by the detector. On the QCD plot it can be seen that after a large count spike at approximately zero, the rest of the counts taper off to zero quickly, with a range of 0 to around 1.1. On the Higgs plot, there are two spikes, one around 0, and another one is around 0.5, which is much bigger.
